# 1번 문제
* 1) 파일 s.txt내의 다음 내용과 함께 아래 문제의 코드를 제시하시오.

```python
file = open("s.txt",'w')
file.write('pig ham\n')
file.write('cat dog\n')
file.write('ham bird\n')
file.write('dog pig\n')
file.close()
```

s.txt로 파일을 열어 file.write를 이용해 한 줄씩 s.txt에 내용을 입력한다.

* 2)파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬 후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오. 

```python
with open('s.txt') as f:
    lines = f.read().splitlines()
    lines.sort()

with open('s1.txt','w',encoding='UTF-8') as f2:
    for i in lines:
        f2.write(i+'\n')
```

.splitlines()를 이용해 파일안에 있는 내용을 한 줄 씩 읽어 리스트화 시킨다.       
읽어들여 리스트화 된 lines를 정렬시킨다.(맨 앞문자 기준으로 정렬이기에 추가적인것 없이 .sort()를 활용함)    
     
`with open(파일겅로, 모드) as 파일객체`를 사용해 파일을 열고 리스트의 요소들을 새로운 파일에 써준다.      
for 문을 사용해 요소마다 줄바꿈을 해주어 각 줄을 다르게 넣어준다.     

* 3) 파일 s.txt를 읽어서 각 라인에 있는 두 번쨰 단어(문자열)자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오

```python
with open('s.txt') as f3:
    lines = f3.read().splitlines()

r_lines = sorted(sorted(lines),key=lambda x:x[4])

with open('s2.txt','w',encoding='UTF-8') as f3:
    for i in r_lines:
        f3.write(i+'\n')
```

위 2번과 동일하게 .splitlines()를 이용해 파일 안에 있는 내용을 한 줄 씩 읽어 리스트화 시킨다.      
그리고 이때는 두번째 단어를 기준으로 정렬을 해야하기 때문에 두 번재 단어들은 인덱스 값 4자리에 모두 위치해 있기 때문에 4번째 요소들을 key값으로 지정해서 이를 기준으로 정렬해준다.       
그리고 정렬한 값을 다시 2번과 동일한 방법으로 s2.txt에 저장을 해 준다.

* 4) 파일 s.txt를 읽어서 각 라인에 있는 모든 단어들을 순차적으로 다시 나열하되 각 라이네 세게의 단어들이 오도록 하여 s3.txt에 기록하는 코드를 작성하시오.

```python
with open('s.txt') as f4:
    lines = f4.read().splitlines()

full_list = ' '.join(lines)

separate_list=full_list.split()

count=0

with open('s3.txt','w',encoding='UTF-8') as f4:
    for i in separate_list:
        f4.write(i+' ')
        count +=1
        if count%3==0:
            f4.writelines("\n")
```

위 2번과 동일하게 이 또한 splitlines()를 통해 라인별로 파일에서 값을 읽어 리스트화 해준다.      
그리고 full_list라는 이름으로 lines라는 리스트를 join을 통해 str로 만들어 준다.    
이후 separate_list라는 이름으로 모은 str을 띄워쓰기 단위로 split해주어 각각의 단어를 요소로 가지는 리스트를 만들어준다.     
count라는 변수를 0으로 초기화 한다.
2번과 동일한 방법으로 s3.txt에 써 저장한다.      
그러나 3개씩 끊어 저장하기 위해 단어 하나를 저장할때마다 count를 1씩 증가시켜 count가 3이 되면 즉 3개의 단어가 한 줄에 써지만 줄바꿈을 저장해 다음 줄에 이어서 저장되게 한다.

# 3번 문제
* range()함수와 유사한 frange()함수를 다음 조건을 참고하여 만드시오.
> 1)  frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
> 2)frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.

```python
import numpy as np


def frange(start, stop=-1, step=0.1):
    result = []
    if stop == -1:  # frange(stop)
        stop = start
        start = 0.0
    lteration = np.arange(start, stop, step)
    for i in lteration:
        result.append(round(i,2))
    return result

print(frange(0.5))
print(frange(1.0,2.0))
print(frange(2.2,4.0,0.5))
```

numpy모듈의 arange함수
> 이는 반열린구간 [ start,stop)에서 step의 크기만큼 일정하게 떨어져 있는 숫자들을 array형태로 반환해 주는 함수이다.      

그러나 frange의 매개변수가 하나가 들어올때도 두개가 들어올때도 있기 때문에 stop과 step은 기본값을 미리 갖게 한다.    
frange에 매개변수가 1개 들어온 경우에는 start값이 아닌 stop값이 들어온 것이므로, 
>if문을 사용해 먄약 stop이 -1인 `매개변수가 하나만 들어와 stop값이 정의가 되지 않았을` 경우: stop값을 매개변수로 입력받은 start값으로 바꿔줘고 start값은 다시 0.0으로 초기화 해준다.     

매개변수가 2개나 3개가 들어왔을 경우에는 특별하게 바뀔 부분은 없다.      

그리고 이를 실질적으로  처리해 주는 부분으로는 numpy모듈의 arange함수를 사용한다.       
lteration에  np.arange(start, stop, step)를 대입해 for문을 lteration동안 실행시키며 처리된 값들을 result라는 리스트에 append 시킨다.   
이때 round함수를 사용해 값을 반올림 시킨다.

# 4번 문제
* 가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

```python
def sum(*args):
    result =0
    for i in args:
        result += i
    return result

print(sum())
print(sum(1,2))
print(sum(1,2,3,4,5))
print(sum(1,5,7,2,-10))
```

가변인수를 받는 sum()함수를 구현한다. *args를 사용해 가변인수를 받고 result라는 변수를 만들어 값을 여기에 계속 더해준다.       
for문을 통해 입력받은 변수들을 모두 더하게 해준다.   


# 5번 문제
* 여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinital()함수를 다음 조건을 참고하여 작성하시오.

```python
def myinital(str):
    str_list = str.split(' ')

    def change_fir(str):
        c_str = str.capitalize()
        return c_str[0]

    alp = list(map(change_fir,str_list))
    result = ''.join(alp)
    print(result)

myinital("as soon as possible")
```

myinital함수를 만들기위해 먼저 매개변수로 입력받은 문자열 str을 .split를 이용해 띄워쓰기 단위로 리스트에 저장한다.     
map함수를 이용하기 위해 새로운 함수를 하나 더 만든다.       
> change_fir이라는 함수는 맨 앞글자를 대문자로 만들어주고 0번째 요소 즉 대문자화 된 첫번재 글자를 return해주는 함수이다.      

alp이라는 변수에 map함수를 이용해 str_list의 요소들을 모두 chage_fir함수를 거치게 한다.     
alp에 반환된 문자들은 result에 join을 통해 문자열로 저장이 된다.


# 6번 문제
* 음이 아닌 정수n을 입력받으면 n!을 계산하는 myfact()함수를 재귀적 함수로 구현하시오.

```python
def myfact(n):
    if n > 1:
        return n*myfact(n-1)
    else:
        return 1

print(myfact(2958))
```

재귀함수를 사용한다.      
만약 입력받은 num이 1보다 크다면 num * myfact(num-1)으로 함수본인을 다시 불러와 곱셈을 계속 진행한다.      
그러나 num이 1보다 크지 않다면 1을 반환해 연산을 끝낸다.

최대로 num(즉 n)을 늘려보니 2958까지 가능함을 알 수 있었다.

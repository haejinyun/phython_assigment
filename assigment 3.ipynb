# 1번 문제
* L=[1, 2, 3, 4, 5]일 때 다음 각 문장을 수행한 후의 결과를 보고 납득할 만한 이유를 설명하시오.


```python
L=[1, 2, 3, 4, 5]
L[1:3] = [100]
print(L)
```

리스트는 ***가변타입*** 으로 항목에 바로 값을 대입할 수 있다.    
고로 위 식은 [2,3]을 삭제하고 [100]을 삽입한다.

```python
L=[1, 2, 3, 4, 5]
L[1:3] = [(100)]
print(L)
```

위 문제와 동일한 문제이다.    
이유로는 [(100)] 과 [100]은 ***동일***하기 때문이다.   
[(100)]의 안의 가로는 ***튜플이 아닌***, 그저 우선 연산자의 ***소괄호***로 인식하기 때문이다.

```python
L=[1, 2, 3, 4, 5]
L[1:3] = 100
print(L)
```

***에러***가 발생한다.      
이유로는, 조각 안에 리스트를 대입할때, ***'대입하려고 하는 대상은 항목이 전혀 없거나'***, ***'하나만 있더라도 반드시 다른 리스트나 컬렉션'***이어야 한다.


```python
L=[1, 2, 3, 4, 5]
L[1:3] = (100)
print(L)
```

***에러***가 발생한다.    
이유로는 위의 변경하려는 (100)은 튜플형태가 아닌, 그저 ***우선 연산자 '소괄호'가 씌워진 '정수 100'***이기에 위의 예제와 같은 원인으로 같은 에러를 발생한다.

```python
L=[1, 2, 3, 4, 5]
L[1:3] = (100, 101, 102)
print(L)
```

슬라이싱 한 부분과 크기가 다르지만 ***튜플은 삽입이 가능***하다.

```python
L=[1, 2, 3, 4, 5]
L[1:3] = [100, 101, 102]
print(L)
```

위의 문제와 같은 방식으로 리스트도 슬라이싱 한 부분과 크기가 달라도 ***리스트 형태 또한 삽입 가능***하다.

```python
L=[1, 2, 3, 4, 5]
L[1:3] = [(100, 101, 102)]
print(L)
```

위 예제를 통해 튜플은 슬라이싱 한 부분과 크기가 달라도 상관이 없다는 것을 보였다.    
[(100, 101, 102)]는 ***리스트로 크기가 하나***이다.     
`0번째 인덱스의 값이 튜플이고 이 튜플은 3개의 요소를 가진다`는 것이다.     
고로 결과는 L 리스트에 1~2까지의 인덱스의 요소값들이 지워지고, (100, 101, 102)튜플을 하나의 리스트 요소로써 가진다.

```python
L=[1, 2, 3, 4, 5]
L[1:3] = [[100, 101, 102]]
print(L)
```

위 문제의 해석과 아주 유사하다.     
위의 L리스트는 튜플을 요소로 가졌다면 이 문제에서는 ***리스트를 요소로 가진 리스트이다.***     
고로 결과는 L 리스트에 1~2까지의 인덱스의 요소값들이 지워지고, [100, 101, 102]리스트를 하나의 리스트 요소로써 가진다.

# 5번 문제
* 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.     
> 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)       
> 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
    
[참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)     
[참고 2]: sum() 등의 내장함수 사용 가능함

```python
# for ~ in 방식
def addall(l): 
    sum=0
    for i in l:
        sum += i
    return sum

addall([1])
```

```python
# 리스트 내포 방식
def addall(l):
    return sum([i for i in l])

print(addall([1,2,3,4,5,6,7,8,9]))
```

##### 1.) for ~ in 사용 방식    
>* 매개변수로 입력받은 리스트 ㅣ을 for in리터럴을 사용해 sum이라는 변수에 리스트 값을 모두 더해준다.


##### 2.)
>* 매개변수로 입력받은 리스트 l을 리스트 내포 ***(i for i in l)*** 를 사용한다.

# 5번 문제(2)

```python
# for ~ in 방식
def addall(l):
    sum=0
    for i in l:
        if(i%2==1):
            sum += i
    return sum

addall([1])
```

```python
# 리스트 내포 방식
# 리스트 내포 방식
def addallodd(l):
    return sum([i for i in l if i%2==1])

print(addallodd([1,2,3,4,5,6,7,8,9]))
```

##### 1.) for ~ in 사용 방식    
>* 매개변수로 입력받은 리스트 ㅣ을 i라는 변수를 사용해 for in리터럴을 사용해 요소들을 확인한다.    
>* 이때 만약 i번째 요소가 2로 나누었을때 나머지 값이 1 인 홀수라면 이때 sum이라는 변수에 리스트 값을 더해줌으로 홀수의 합을 찾는다.


##### 2.)
>* 매개변수로 입력받은 리스트 l을 리스트 내포 ***(i for i in l if ~)*** 를 사용한다.   
>* 이때 리스트 내포에 if 문을 붙여 i번째 요소가 2로 나누었을때 나머지 값이 1인 홀수라면 sum에 바로 더할 수 있게 한다.

# 7번 문제
* 사전 d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}가 주어졌을 때 다음 요구사항에 맞는 코드를 제시하시오     
* [참고]: d.keys(), d.values()를 통해 리스트를 얻어낸 후 리스트가 지니고 있는 sort(cmp=None, key=None, reverse=False)함수를 활용하시오.    
> 1) 키의 알파벳 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.     
> 2) 키의 알파벳 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.     
> 3) 값의 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.     
> 4) 값의 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.        

```python
import operator

d = {'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5}

s_dk = sorted(d.items()) #key를 기준으로 오름차순
print(s_dk)

r_dk = sorted(d.items(), reverse=True) #key를 기준을 내림차순
print(r_dk)

s_dv = sorted(d.items(), key=operator.itemgetter(1))
print(s_dv)

r_dv = sorted(d.items(), reverse=True, key=operator.itemgetter(1))
print(r_dv)
```

#### d라는 사전을 정렬을 하고자 한다면, sorted()를 사용한다.   

>* ***`s_dk = sorted(d.items())`*** 는 딕셔너리 d의 키와 값의 쌍이 키를 기준으로 정렬되어 튜플 자료형으로 s_dk로 들어간다.    

>* ***`r_dk = sorted(d.items(), reverse=True)`*** 는 위 s_dk와 동일하지만 sorted함수에 ***reverse=True***가 들어가 내림차순으로 정렬 되어 튜플 자료형으로 r_dk에 들어간다.      

>* ***`s_dv = sorted(d.items(), key=operator.itemgetter(1))`*** 는 값을 기준으로 오름차순을 하는 것으로 ***key=operator.itemgetter(1)*** 를 sorted함수에 추가해 value값을 기준으로 정렬 할 수 있다.    

>* ***`r_dv = sorted(d.items(), reverse=True, key=operator.itemgetter(1))`*** 는 값을 기준으로 내린차순을 하는 것으로 r_dk와 s_dv를 섞어 만들었다. ***reverse=True***를 통해 내림차순으로 정렬되게 하고, ***key=operator.itemgetter(1)*** 를 통해 value값으로 정렬을 할 수 있게 했다.

# 8번 문제
* 다음 문제를 정확히 읽고 적합한 코딩을 제시하시오.
>* length(X)를 다음과 같이 정의한다 : 10진수로 표현된 X의 자릿수    
>* Strange라는 숫자를 아래과 같이 정의한다.     
>>* 음이 아닌 정수 중 자릿수가 1인 모든 숫자는 Strange이다.      
>>- length(X) > 1 인 숫자 X 는 다음 2가지를 만족하면 Strange이다.     
>>> * X는 length(X)로 정확히 나누어 떨어진다.      
>>>* X / length(X) 가 strnage이다.      

>* 이 문제에서 숫자 하나가 주어질 때, 이 숫자가 Strange인지 판단하는 함수 isStrange(num)을 작성하시오.      
>* isStrange(num) 함수가 재귀적인 함수가 되도록 작성해 보세요.       

```python
def isStrange(num):
    length = lambda x: len(str(x))
    if 0 <= num:
        if 0 <= length(num) <= 1:
            return True
        elif 1 < length(num):
            x = (int)(num / length(num))
            if (num % length(num) == 0) and (isStrange(x)):
                return True
            else:
                return False
    else:
        return False


print(isStrange(0))
print(isStrange(56))
print(isStrange(100))
print(isStrange(7882602506492313600))
print(isStrange(6012302549522098651))

```

람다 함수를 사용해 length(x)를 정의함.   
> 이는 입력받은 숫자의 자릿수 즉 길이를 반환하는 것.  

만약 입력받은 숫자가 ***음수가 아닐때***, 
>숫자의 length()의 반환 값이 ***0혹은 1*** 이라면 ***True를 반환***한다.(고로 이 수는 strange이다.)     
>숫자의 length()의 반환 값이 ***1 초과***라면 and연산자를 통해 아래 두가지의 확인 절차를 거친다.        
>>* 입력받은 숫자 ***num이 num의 자리수 인 length(x)로 나누어 떨어지***는지 확인을 한다.

# 9번 문제
* 다음 문제를 정확히 읽고 적합한 코딩을 제시하시오.

>* 1번 그룹, 2번 그룹, ... n번 그룹으로 번호가 매겨진 총 n개의 그룹이 존재하고, ith 그룹에는 임의의 팀원 수 ai 명이 있다.
하나의 버스 정류장이 있고, 버스는 오직 한 대만 운영한다. 버스는 버스 정류장에 도착하고 모든 그룹은 버스를 통해 여행하길 원한다.
하지만, 같은 그룹에 속한 팀원들은 분리되기를 원치 않는다. 그래서 그룹의 모든 팀원이 탈 수 있을 때 에만 그 그룹은 버스에 탑승한다.
또한, 이들은 그룹의 번호순으로 여행을 하길 원한다. 예를 들어, 3번 그룹은 1번 그룹과 2번 그룹이 여행을 한 뒤에 버스를 탈 수 있거나, 1번 그룹과 2번 그룹이 이미 버스에 탑승한 상태에 있을 때 동시에 버스를 탈 수 있다.
마지막으로, 버스는 빈 좌석이 없을 때 에만 여행을 떠난다.

>* 이 문제에서 각 그룹의 팀원 수가 리스트로 주어질 때, 가능한 모든 버스의 사이즈 x(x명의 사람들이 동시에 탈 수 있음)를 반환하는 함수 solve(a) 를 작성하시오.
>* [주의] 반드시 계산 복잡도를 최대한 낮추세요. 아래 예시 3번을 실행했을 때 결과가 바로 나타날 수 있도록!!

```python
a = [1, 2, 1, 1, 1, 2, 1, 3]
b = [1, 1, 1, 1]
c = [9999] * 10 ** 5



def solve(a):
    sum = 0
    check_sum = 0
    b = list(set(a))
    max = b[-1]  # 주어진 리스트에서 가장 큰 값을 b에 지정(가장 인원이 많은 그릅의 인원 수)
    for i in range(0, len(a)):  # 모든 그룹의 인원을 더하는 for문
        sum += a[i]
    l = []

    for i in range(1, int(sum ** 0.5) + 1):
        if sum % i == 0:
            l.append(i)
            if ((sum // i) != i):
                l.append(sum // i)

    r_l = sorted(l)

    i=0
    while i < len(r_l):  # 리스트에 max 값보다 작은 자식을 너굴맨이 처리했다구..
        if r_l[i] < max:
            del r_l[i]
        else:
            i += 1  # 리스트 값이 밀려서 처리 안되는 경우가 생기기 때문에

    k = 0
    check_l = []
    while k < len(r_l):
        j = 0
        while j < len(a):
            check_sum += a[j]
            if check_sum < r_l[k]:
                j+=1
            elif check_sum == r_l[k]:
                check_sum = 0
                j+=1
            else:
                break
        if j == len(a):
            check_l.append(r_l[k])
        k+=1
        check_sum=0
    print(check_l)

solve(a)
print()
solve(b)
print()
solve(c)
```

문제를 해결한 알고리즘으로는 a 리스트 즉 n개의 그룹의 명수를 모두 더해 총 인원을 알아내고, 그 인원들의 약수를 구한다.
이 약수들이 a 리스트에서 탈 수 있는 n인용 버스 사이즈이다. 그러나 이 버스는 모두 가능하지 않다. 예외를 제하며 탙 수 있는 버스 사이즈를 구한다.

>위의 가정을 살펴보면 이는 약수로 1, 2, 7, 14인용이 가능하지만 가장 큰 그룹의 인원 수가 5이므로 7,14인용 버스만 가능하다.

# Incremental Project 문제

```python
import requests
import re
import string

req = requests.get('https://www.koreatech.ac.kr')

source = req.text

def delete(s):
    body = re.search('<body.*/body>', s, re.I | re.S)
    body = body.group()
    body = re.sub('<script.*?>.*?</script>', '', body, 0, re.I | re.S)
    body = re.sub('<style.*?>.*?</style>', '', body, 0, re.I | re.S)
    body = re.sub('<url.*?>.*?</url>', '', body, 0, re.I | re.S)
    text = re.sub('<.+?>', '', body, 0, re.I | re.S)
    nospace = re.sub('&nbsp;| |\t|\r|', '', text)

    cleaned_text = text.translate(str.maketrans('', '', string.punctuation))  # 구두점 지우기
    word_list = cleaned_text.split()
    count = {}
    for i in word_list:
        try:
            count[i] += 1
        except:
            count[i] = 1
    return count

print(delete(source))
```

정규 표현식을 통해 html과 script, css, style문을 지웠다.     
search() 메소드의 첫번째 인수 `'<body.*</body>'` 는 <body로 시작하고 어떤 글자가(.) 없을 수도 여러개 있을수도 (*) 있고 그 이후에 </body>로 끝난다는 정규 표현식아다. 그 뒤 `re.I는 대소문자를 구분하지 않는다는 의미이고`, `| re.S`는 여러 줄에 걸쳐서 정규 표현식을 찾으라는 의미이다.     

`<body> 안의 <script>...</script> 삭제`를 위해 sub()함수를 사용한다.     
`'<script.*?>.*?</script>'`의 의미는 <script로 시작하고 그 뒤에 어떤 글자가 없을 수도 있고 여러개 있을 수도 있으며 (*? - ex, src='...'), 그 뒤에는 >가 나오고 다시 어떤 내용이 있을 수 있으며 끝은 `</script>`로 끝난다는 내용이다.      
>이로 스크립트 문에 지워진다.

위의 스크립트문을 지울때와 동일하게 `style문`과 `url문`을 지워준다.       
또한 테그 및 주석을 지우기위해 `sub('<.+?>', '', body, 0, re.I | re.S)`함수를 사용한다.      
마지막으로 공백을 지우기 위해 `sub('&nbsp;| |\t|\r|', '', text)`함수를 사용한다.

구두문자를 지우기 위해 string.punctuation을 사용한다.     
이는 구두점으로 간주되는 모든 문자를 포함하고 있다.     
String클래스에 제공된 내장 함수를 사용해 구두점을 지운다.    

중복 요소를 카운팅 하고 dict로 표현하기 위해 예외 처리를 하는 식으로 ***try except***를 사용한다.     
***for 문***을 통해 word_list의 요소를 하나씩 꺼내 ***count라는 이름의 딕셔너리에 넣는다.***   
이때 `이미 존재하는 key값이라면 try문이 실행되며 vlaue에 +1`이 된다.       
`아니라면 except가 실행되며 value값은 기존 1을 유지`함으로 표현했다.
